```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import pickle
```

```python
df_c = pd.read_csv("./resumendatos.csv")
H_op = df_c.K_operacion_GWs.to_numpy() / 25
df_c["H_operacion"] = H_op
Ta = []
Tb = []
Tc = []
Td = []
H  = []
Kd = []
for fecha in df_c.fecha:
    path = "./2023_07_27/model_c/"+fecha+"_sampled_result.p"
    file = open(path, "rb")
    params = pickle.load(file)
    Ta.append(params.x[0])
    Tb.append(params.x[1])
    Tc.append(params.x[2])
    Td.append(params.x[3])
    H.append(params.x[4])
    Kd.append(params.x[5])
    file.close()
df_c["Ta"] = Ta
df_c["Tb"] = Tb
df_c["Tc"] = Tc
df_c["Td"] = Td
df_c["H"] = H
df_c["Kd"] = Kd
df_c
```
